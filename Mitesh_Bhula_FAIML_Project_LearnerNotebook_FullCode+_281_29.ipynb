## **Data Description**

The data is about customers of the treadmill product(s) of a retail store called Cardio Good Fitness. It contains the following variables-


* Product - The model no. of the treadmill
* Age -  Age of the customer in no of years
* Gender - Gender of the customer
* Education - Education of the customer in no. of years
* Marital Status - Marital status of the customer
* Usage - Avg. number of times the customer wants to use the treadmill every week
* Fitness - Self rated fitness score of the customer (5 - very fit, 1 - very unfit)
* Income - Income of the customer
* Miles- Miles that a customer expects to run

### **Bivariate and Multivariate Analysis**

### **Question 7:** Perform bivariate and multivariate analysis to explore relationships between the important variables in the dataset. (It is a good idea to explore relations between numerical variables as well as relations between numerical and categorical variables) [12 marks]

```python
# Write your code here
# Using a countplot to determine the relationship between gender and product
sns.countplot(data=cardio, x='Product', hue='Gender');
```

```python
# Fitness level in respect to user age
sns.lineplot(data=cardio, x='Fitness', y='Age', hue='Gender', ci=False);
```

```python
# Age effect on Usage
sns.jointplot(data=cardio, x='Usage', y='Age', kind='hex');
plt.colorbar();
```

```python
# Fitness effect on Usage
sns.jointplot(data=cardio, x='Usage', y='Fitness', kind='hex');
plt.colorbar();
```

```python
# Checking correlation between numerical variables
numerical_values = ['Age', 'Education', 'Usage', 'Fitness', 'Income', 'Miles']
correlation = cardio[numerical_values].corr()
# Plotting heatmap
plt.figure(figsize=(15, 7))
sns.heatmap(correlation, annot=True, vmin=-1, vmax=1, fmt=".2f", cmap="Spectral")
plt.show()
```

```python
sns.pairplot(data=df[numerical_values], diag_kind="kde")
plt.show()
```

```python
# Relationship between usage and each different product
plt.figure(figsize=(15,7))
sns.boxplot(x=df['Product'], y=df['Usage'])
plt.ylabel('Usage')
plt.xlabel('Products')
plt.show()
```

```python
# Dispersion of age and usage
sns.catplot(x='Age', col='Usage', data=df, col_wrap=4, kind="violin")
plt.show()
```

```python
# Relationship between product use, gender and miles run
#sns.catplot(x = 'Product', y = 'Gender', data=cardio, kind = 'bar')
sns.catplot(x='Product', y='Miles', data=cardio, kind="bar", hue='Gender')
plt.show()
```

### Observations: [Write your observation(s) here]

The TM195 product is used equally by both genders, whereas the TM789 is used by more Males than females.

The females fitness decreases as age increases, whereas the Males fitness levels increase as age increases.

Users aged around 25 years old used the treadmills 3 or more times per week.

The correlation between Fitness and Usage is as we expected, the fitter a user is, the more the more they will use the treadmills.

As per the pairplot, we see a positive correlation between age and income, depicting as users get older their income may increase.

The pair plot also shows us a 0.04 correlation between Miles and Age, this means that there are no specific age groups that run more miles than others. The data is spread evenly across the sample.

The violion plot shows us the dispersion of age for the number of times a user utilises the treadmiles. We can see as age increases the number of users decreases.

We notice users run the most miles on the TM798, more females use this product than men. The females also run more miles on this machine than males. The most popular machine from the data, TM195, is the least used product, this may indicate users are not comfortable using the new product.
